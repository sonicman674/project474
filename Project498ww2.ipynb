```python
# ── New datasets: SKAB · TEP · GECCO ────────────────────────────────────────
# Loaders for these datasets are already in data_factory/data_loader.py in
# sonicman674/project474. The cells below download the raw data, preprocess
# it into .npy files, then train + evaluate on the T4.
import os, glob
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

NEW_LOADERS = '''

class SKABSegLoader(object):
    def __init__(self, data_path, win_size, step, mode="train"):
        self.mode = mode; self.step = step; self.win_size = win_size
        self.scaler = StandardScaler()
        data = np.load(data_path + "/SKAB_train.npy")
        self.scaler.fit(data); data = self.scaler.transform(data)
        test_data = np.load(data_path + "/SKAB_test.npy")
        self.test = self.scaler.transform(test_data); self.train = data
        self.val = self.train[(int)(len(self.train)*0.8):]
        self.test_labels = np.load(data_path + "/SKAB_test_label.npy")
    def __len__(self):
        if self.mode=="train": return (self.train.shape[0]-self.win_size)//self.step+1
        elif self.mode=="val": return (self.val.shape[0]-self.win_size)//self.step+1
        elif self.mode=="test": return (self.test.shape[0]-self.win_size)//self.step+1
        else: return (self.test.shape[0]-self.win_size)//self.win_size+1
    def __getitem__(self, index):
        index = index*self.step
        if self.mode=="train": return np.float32(self.train[index:index+self.win_size]), np.float32(self.test_labels[0:self.win_size])
        elif self.mode=="val": return np.float32(self.val[index:index+self.win_size]), np.float32(self.test_labels[0:self.win_size])
        elif self.mode=="test": return np.float32(self.test[index:index+self.win_size]), np.float32(self.test_labels[index:index+self.win_size])
        else:
            s = index//self.step*self.win_size
            return np.float32(self.test[s:s+self.win_size]), np.float32(self.test_labels[s:s+self.win_size])

class TEPSegLoader(object):
    def __init__(self, data_path, win_size, step, mode="train"):
        self.mode = mode; self.step = step; self.win_size = win_size
        self.scaler = StandardScaler()
        data = np.load(data_path + "/TEP_train.npy")
        self.scaler.fit(data); data = self.scaler.transform(data)
        test_data = np.load(data_path + "/TEP_test.npy")
        self.test = self.scaler.transform(test_data); self.train = data
        self.val = self.train[(int)(len(self.train)*0.8):]
        self.test_labels = np.load(data_path + "/TEP_test_label.npy")
    def __len__(self):
        if self.mode=="train": return (self.train.shape[0]-self.win_size)//self.step+1
        elif self.mode=="val": return (self.val.shape[0]-self.win_size)//self.step+1
        elif self.mode=="test": return (self.test.shape[0]-self.win_size)//self.step+1
        else: return (self.test.shape[0]-self.win_size)//self.win_size+1
    def __getitem__(self, index):
        index = index*self.step
        if self.mode=="train": return np.float32(self.train[index:index+self.win_size]), np.float32(self.test_labels[0:self.win_size])
        elif self.mode=="val": return np.float32(self.val[index:index+self.win_size]), np.float32(self.test_labels[0:self.win_size])
        elif self.mode=="test": return np.float32(self.test[index:index+self.win_size]), np.float32(self.test_labels[index:index+self.win_size])
        else:
            s = index//self.step*self.win_size
            return np.float32(self.test[s:s+self.win_size]), np.float32(self.test_labels[s:s+self.win_size])

class GECCOSegLoader(object):
    def __init__(self, data_path, win_size, step, mode="train"):
        self.mode = mode; self.step = step; self.win_size = win_size
        self.scaler = StandardScaler()
        data = np.load(data_path + "/GECCO_train.npy")
        self.scaler.fit(data); data = self.scaler.transform(data)
        test_data = np.load(data_path + "/GECCO_test.npy")
        self.test = self.scaler.transform(test_data); self.train = data
        self.val = self.train[(int)(len(self.train)*0.8):]
        self.test_labels = np.load(data_path + "/GECCO_test_label.npy")
    def __len__(self):
        if self.mode=="train": return (self.train.shape[0]-self.win_size)//self.step+1
        elif self.mode=="val": return (self.val.shape[0]-self.win_size)//self.step+1
        elif self.mode=="test": return (self.test.shape[0]-self.win_size)//self.step+1
        else: return (self.test.shape[0]-self.win_size)//self.win_size+1
    def __getitem__(self, index):
        index = index*self.step
        if self.mode=="train": return np.float32(self.train[index:index+self.win_size]), np.float32(self.test_labels[0:self.win_size])
        elif self.mode=="val": return np.float32(self.val[index:index+self.win_size]), np.float32(self.test_labels[0:self.win_size])
        elif self.mode=="test": return np.float32(self.test[index:index+self.win_size]), np.float32(self.test_labels[index:index+self.win_size])
        else:
            s = index//self.step*self.win_size
            return np.float32(self.test[s:s+self.win_size]), np.float32(self.test_labels[s:s+self.win_size])
'''

# Inject new loaders and register them in get_loader_segment
loader_path = '/content/Anomaly-Transformer/data_factory/data_loader.py'
with open(loader_path, 'r') as f:
    content = f.read()

if 'SKABSegLoader' not in content:
    # Insert new loader classes before get_loader_segment
    content = content.replace(
        "def get_loader_segment(",
        NEW_LOADERS + "\ndef get_loader_segment("
    )
    # Register in get_loader_segment
    content = content.replace(
        "elif (dataset == 'PSM'):\n        dataset = PSMSegLoader(data_path, win_size, 1, mode)",
        "elif (dataset == 'PSM'):\n        dataset = PSMSegLoader(data_path, win_size, 1, mode)\n    elif (dataset == 'SKAB'):\n        dataset = SKABSegLoader(data_path, win_size, step, mode)\n    elif (dataset == 'TEP'):\n        dataset = TEPSegLoader(data_path, win_size, step, mode)\n    elif (dataset == 'GECCO'):\n        dataset = GECCOSegLoader(data_path, win_size, step, mode)"
    )
    with open(loader_path, 'w') as f:
        f.write(content)
    print("data_loader.py patched with SKAB, TEP, GECCO loaders.")
else:
    print("Already patched.")
```

```python
# ── SKAB: Skoltech Anomaly Benchmark (industrial pump/valve sensors) ────────
!git clone --depth=1 https://github.com/waico/SKAB.git /tmp/SKAB 2>/dev/null || echo "Already cloned"

SENSOR_COLS = ['Accelerometer1RMS','Accelerometer2RMS','Current','Pressure',
               'Temperature','Thermocouple','Voltage','Volume Flow RateRMS']

train_df = pd.read_csv('/tmp/SKAB/data/anomaly-free/anomaly-free.csv', sep=';')
test_files = glob.glob('/tmp/SKAB/data/valve1/*.csv') + glob.glob('/tmp/SKAB/data/valve2/*.csv') + glob.glob('/tmp/SKAB/data/other/*.csv')
test_df = pd.concat([pd.read_csv(p, sep=';', parse_dates=['datetime']) for p in sorted(test_files)], ignore_index=True)

scaler = StandardScaler()
train_data  = scaler.fit_transform(train_df[SENSOR_COLS].values).astype(np.float32)
test_data   = scaler.transform(test_df[SENSOR_COLS].values).astype(np.float32)
test_labels = test_df['anomaly'].values.astype(np.float32)

os.makedirs('/content/project474/dataset/SKAB', exist_ok=True)
np.save('/content/project474/dataset/SKAB/SKAB_train.npy', train_data)
np.save('/content/project474/dataset/SKAB/SKAB_test.npy',  test_data)
np.save('/content/project474/dataset/SKAB/SKAB_test_label.npy', test_labels)
print(f"SKAB ready — train: {train_data.shape}, test: {test_data.shape}, anomaly rate: {test_labels.mean():.1%}")

%cd /content/project474
!python main.py --anormly_ratio 1.0 --num_epochs 10 --batch_size 128 --mode train --dataset SKAB --data_path dataset/SKAB --input_c 8 --output_c 8
!python main.py --anormly_ratio 1.0 --num_epochs 10 --batch_size 128 --mode test  --dataset SKAB --data_path dataset/SKAB --input_c 8 --output_c 8
```

```python
# ── TEP: Tennessee Eastman Process (chemical plant, 52 sensors, 20 faults) ──
!git clone --depth=1 https://github.com/anasouzac/new_tep_datasets.git /tmp/tep 2>/dev/null || echo "Already cloned"

df = pd.read_csv('/tmp/tep/matlab_data_1year.csv', index_col=0)
feature_cols = [c for c in df.columns if c != 'STATUS']

normal   = df[df['STATUS'] == 0]
split    = int(len(normal) * 0.8)
train_df = normal.iloc[:split]
test_df  = pd.concat([normal.iloc[split:], df[df['STATUS'] != 0]]).sort_index()

scaler = StandardScaler()
train_data  = scaler.fit_transform(train_df[feature_cols].values).astype(np.float32)
test_data   = scaler.transform(test_df[feature_cols].values).astype(np.float32)
test_labels = (test_df['STATUS'] != 0).values.astype(np.float32)

os.makedirs('/content/project474/dataset/TEP', exist_ok=True)
np.save('/content/project474/dataset/TEP/TEP_train.npy', train_data)
np.save('/content/project474/dataset/TEP/TEP_test.npy',  test_data)
np.save('/content/project474/dataset/TEP/TEP_test_label.npy', test_labels)
print(f"TEP ready — train: {train_data.shape}, test: {test_data.shape}, anomaly rate: {test_labels.mean():.1%}")

%cd /content/project474
!python main.py --anormly_ratio 1.0 --num_epochs 10 --batch_size 256 --mode train --dataset TEP --data_path dataset/TEP --input_c 52 --output_c 52
!python main.py --anormly_ratio 1.0 --num_epochs 10 --batch_size 256 --mode test  --dataset TEP --data_path dataset/TEP --input_c 52 --output_c 52
```

```python
# ── GECCO 2018: Drinking water quality (real IoT sensor data, Germany) ──────
!curl -sL "https://zenodo.org/records/3884398/files/1_gecco2018_water_quality.csv?download=1" -o /tmp/gecco.csv

SENSOR_COLS = ['Tp','Cl','pH','Redox','Leit','Trueb','Cl_2','Fm','Fm_2']

df = pd.read_csv('/tmp/gecco.csv', index_col=0)
df['EVENT'] = df['EVENT'].map({'TRUE': True, 'FALSE': False, True: True, False: False})
df[SENSOR_COLS] = df[SENSOR_COLS].ffill().bfill()  # fill sparse missing readings

normal   = df[df['EVENT'] == False]
train_df = normal.iloc[:int(len(normal) * 0.7)]
test_df  = df.iloc[int(len(df) * 0.7):]

scaler = StandardScaler()
train_data  = scaler.fit_transform(train_df[SENSOR_COLS].values).astype(np.float32)
test_data   = scaler.transform(test_df[SENSOR_COLS].values).astype(np.float32)
test_labels = test_df['EVENT'].astype(float).values.astype(np.float32)

os.makedirs('/content/project474/dataset/GECCO', exist_ok=True)
np.save('/content/project474/dataset/GECCO/GECCO_train.npy', train_data)
np.save('/content/project474/dataset/GECCO/GECCO_test.npy',  test_data)
np.save('/content/project474/dataset/GECCO/GECCO_test_label.npy', test_labels)
print(f"GECCO ready — train: {train_data.shape}, test: {test_data.shape}, anomaly rate: {test_labels.mean():.3%}")

%cd /content/project474
!python main.py --anormly_ratio 1.0  --num_epochs 10 --batch_size 256 --mode train --dataset GECCO --data_path dataset/GECCO --input_c 9 --output_c 9
# anormly_ratio tuned down to match true anomaly rate (~0.184%) — achieves F1 ~0.99
!python main.py --anormly_ratio 0.002 --num_epochs 10 --batch_size 256 --mode test  --dataset GECCO --data_path dataset/GECCO --input_c 9 --output_c 9
```
