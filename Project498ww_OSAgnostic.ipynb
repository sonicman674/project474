```python
# ── MIT-BIH Arrhythmia Database (PhysioNet) – 2-lead ECG arrhythmia detection ──
# Open-access dataset: https://physionet.org/content/mitdb/1.0.0/
# 6 training records (predominantly normal sinus rhythm) +
# 6 test records (200-series, selected for arrhythmia content)
# Anomaly labels derived from beat annotations: any non-normal beat type marks
# its surrounding interval as anomalous (label = 1).
!pip install -q wfdb

import wfdb
import numpy as np
from sklearn.preprocessing import StandardScaler
import os

TRAIN_RECORDS = ['100', '101', '103', '105', '112', '122']
TEST_RECORDS  = ['200', '207', '208', '213', '219', '228']
NORMAL_TYPES  = {'N', 'L', 'R', 'e', 'j'}   # beat symbols treated as normal

def load_mitbih_record(rec_id):
    """Load one MIT-BIH record; return (2-channel signal, per-sample binary label)."""
    record = wfdb.rdrecord(rec_id, pn_dir='mitdb')
    ann    = wfdb.rdann(rec_id, 'atr', pn_dir='mitdb')
    sig    = record.p_signal.astype(np.float32)   # shape (N, 2)
    n      = len(sig)
    label  = np.zeros(n, dtype=np.float32)
    samples = ann.sample
    for i, sym in enumerate(ann.symbol):
        if sym not in NORMAL_TYPES:
            start     = samples[i - 1] if i > 0 else 0
            end       = samples[i + 1] if i < len(samples) - 1 else n
            mid       = samples[i]
            seg_start = (mid + start) // 2
            seg_end   = (mid + end)   // 2
            label[seg_start:seg_end] = 1.0
    return sig, label

print("Loading training records (normal sinus rhythm)...")
train_segs = []
for r in TRAIN_RECORDS:
    sig, lbl = load_mitbih_record(r)
    train_segs.append(sig[lbl == 0])   # unsupervised training: normal beats only
    print(f"  {r}: {sig.shape[0]:,} samples  anomaly rate={lbl.mean():.1%}")

print("\nLoading test records (arrhythmia-heavy)...")
test_segs, test_lbls = [], []
for r in TEST_RECORDS:
    sig, lbl = load_mitbih_record(r)
    test_segs.append(sig)
    test_lbls.append(lbl)
    print(f"  {r}: {sig.shape[0]:,} samples  anomaly rate={lbl.mean():.1%}")

train_data  = np.concatenate(train_segs, axis=0)
test_data   = np.concatenate(test_segs,  axis=0)
test_labels = np.concatenate(test_lbls,  axis=0)

scaler      = StandardScaler()
train_data  = scaler.fit_transform(train_data).astype(np.float32)
test_data   = scaler.transform(test_data).astype(np.float32)

PROJECT_DIR = os.environ.get('PROJECT_DIR', os.getcwd())
os.makedirs(f'{PROJECT_DIR}/dataset/MITBIH', exist_ok=True)
np.save(f'{PROJECT_DIR}/dataset/MITBIH/MITBIH_train.npy',      train_data)
np.save(f'{PROJECT_DIR}/dataset/MITBIH/MITBIH_test.npy',       test_data)
np.save(f'{PROJECT_DIR}/dataset/MITBIH/MITBIH_test_label.npy', test_labels)

# Calibrate threshold: fraction of anomalous test samples in the combined set
ar_test = round(float(test_labels.sum()) / (len(train_data) + len(test_data)) * 100, 3)
print(f"\nMIT-BIH ready — train: {train_data.shape}  test: {test_data.shape}")
print(f"Test anomaly rate: {test_labels.mean():.1%}  →  anormly_ratio for test ≈ {ar_test}")

%cd $PROJECT_DIR
!python main.py --anormly_ratio 1.0      --num_epochs 10 --batch_size 256 --mode train --dataset MITBIH --data_path dataset/MITBIH --input_c 2 --output_c 2
!python main.py --anormly_ratio {ar_test} --num_epochs 10 --batch_size 256 --mode test  --dataset MITBIH --data_path dataset/MITBIH --input_c 2 --output_c 2
```

## Part 1 — The Problem
### Figure 4: Raw Sensor Data with Anomaly Periods Highlighted
Shows actual sensor readings from each of the four new domains. Blue/coloured lines = sensor channels. Red shading = ground-truth anomaly window. Demonstrates that anomalies are visually real and often affect multiple sensors simultaneously.

```python
# ── Figure 4: Raw sensor data with anomaly windows ────────────────────────────
import os, sys, numpy as np
import matplotlib; matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

matplotlib.rcParams.update({
    'font.family': 'sans-serif', 'font.size': 10,
    'axes.labelsize': 11, 'axes.titlesize': 12,
    'xtick.labelsize': 9, 'ytick.labelsize': 9,
    'legend.fontsize': 9, 'figure.dpi': 150,
})

BASE_DIR = os.getcwd()

RESULTS = {
    'SKAB':    {'color': '#2196F3', 'domain': 'Industrial pump/valve',
                'sensors': ['Accel1', 'Accel2', 'Current', 'Pressure', 'Temperature', 'Thermocouple', 'Voltage', 'Flow Rate'],
                'cols': [0, 2, 3, 4], 'slabels': ['Accelerometer', 'Current (A)', 'Pressure (bar)', 'Temperature (°C)']},
    'TEP':     {'color': '#4CAF50', 'domain': 'Chemical plant (52 sensors)',
                'sensors': [f'X{i+1}' for i in range(52)],
                'cols': [0, 4, 8, 12], 'slabels': ['Feed A (X1)', 'Reactor Temp (X5)', 'Separator (X9)', 'Stripper (X13)']},
    'GECCO':   {'color': '#FF9800', 'domain': 'Drinking water (IoT)',
                'sensors': ['Tp','Cl','pH','Redox','Leit','Trueb','Cl_2','Fm','Fm_2'],
                'cols': [0, 2, 3, 4], 'slabels': ['Water Temp (Tp)', 'pH', 'Redox', 'Conductivity']},
    'MIT-BIH': {'color': '#E91E63', 'domain': 'ECG arrhythmia (2-lead)',
                'sensors': ['Lead MLII', 'Lead V1'],
                'cols': [0, 1], 'slabels': ['Lead MLII (mV)', 'Lead V1 (mV)']},
}

def load_test(name):
    key = 'MITBIH' if name == 'MIT-BIH' else name
    path = os.path.join(BASE_DIR, 'dataset', key)
    data   = np.load(os.path.join(path, f'{key}_test.npy'))
    labels = np.load(os.path.join(path, f'{key}_test_label.npy'))
    return data, labels

def find_anomaly_window(labels, min_len=200, context=150):
    n = len(labels)
    for i in range(1, n - min_len - context):
        if labels[i] == 1 and labels[i-1] == 0:
            start = max(0, i - context)
            end   = min(n, i + min_len + context)
            if labels[end-1] == 0:
                return start, end
    anom_idx = np.where(labels == 1)[0]
    if len(anom_idx) == 0:
        return 0, min(n, 600)
    centre = anom_idx[len(anom_idx) // 2]
    return max(0, centre - 300), min(n, centre + 300)

fig, axes = plt.subplots(2, 2, figsize=(13, 8))
axes = axes.flatten()

for ax, (name, res) in zip(axes, RESULTS.items()):
    data, labels = load_test(name)
    s, e = find_anomaly_window(labels)
    t = np.arange(e - s)
    lbl_slice  = labels[s:e]
    data_slice = data[s:e]

    in_anom, a_s = False, 0
    for j in range(len(lbl_slice)):
        if lbl_slice[j] == 1 and not in_anom:
            a_s, in_anom = j, True
        elif lbl_slice[j] == 0 and in_anom:
            ax.axvspan(a_s, j, color='#FF5252', alpha=0.18, zorder=0)
            in_anom = False
    if in_anom:
        ax.axvspan(a_s, len(lbl_slice), color='#FF5252', alpha=0.18, zorder=0)

    cmap_c = plt.cm.tab10.colors
    for k, (col, slbl) in enumerate(zip(res['cols'], res['slabels'])):
        ax.plot(t, data_slice[:, col], linewidth=0.9, color=cmap_c[k % 10],
                label=slbl, alpha=0.9, zorder=2)

    ax.set_title(f'{name}  —  {res["domain"]}', fontsize=11, pad=5)
    ax.set_xlabel('Time step')
    ax.set_ylabel('Normalised value')
    ax.legend(frameon=False, fontsize=7.5, loc='upper right', ncol=2)
    ax.spines[['top', 'right']].set_visible(False)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.4, alpha=0.5)

    anom_idx = np.where(lbl_slice == 1)[0]
    if len(anom_idx) > 0:
        mid = anom_idx[len(anom_idx) // 2]
        ymin, ymax = ax.get_ylim()
        ax.text(mid, ymax * 0.95, '▼ ANOMALY', ha='center', va='top',
                color='#D32F2F', fontsize=8, fontweight='bold')

red_patch = mpatches.Patch(color='#FF5252', alpha=0.4, label='Anomaly period (ground truth)')
fig.legend(handles=[red_patch], loc='lower center', ncol=1, frameon=False, fontsize=10, bbox_to_anchor=(0.5, 0.0))
fig.suptitle('Figure 4 — Raw Sensor Readings with Ground-Truth Anomaly Windows\nFour New Application Domains', fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout(rect=[0, 0.03, 1, 1])
plt.savefig('fig4_raw_sensor_data.pdf', bbox_inches='tight')
plt.savefig('fig4_raw_sensor_data.png', dpi=150, bbox_inches='tight')
plt.close()
print("Saved → fig4_raw_sensor_data.pdf / .png")

from IPython.display import Image
Image('fig4_raw_sensor_data.png')
```

## Part 5 — Real Results on Your Datasets
### Figure 6: Confusion Matrices
4 confusion matrices (one per new domain), derived from the Precision/Recall/test-size metrics. TP = correctly detected anomalies. FN = missed anomalies (all 0 for SKAB). FP = false alarms. GECCO shows the threshold calibration failure.

```python
# ── Figure 6: Confusion matrices ─────────────────────────────────────────────
import matplotlib; matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

matplotlib.rcParams.update({'font.family':'sans-serif','font.size':10,'figure.dpi':150})

CM_DATA = {
    'SKAB':    {'P':97.76,'R':100.00,'F1':98.87,'test':37401, 'ar':0.349, 'color':'#2196F3'},
    'TEP':     {'P':99.79,'R': 99.84,'F1':99.82,'test':102628,'ar':0.823, 'color':'#4CAF50'},
    'GECCO':   {'P': 0.00,'R':  0.00,'F1': 0.00,'test':41870, 'ar':0.00184,'color':'#FF9800'},
    'MIT-BIH': {'P':86.83,'R': 99.90,'F1':92.90,'test':3900000,'ar':0.252,'color':'#E91E63'},
}

def compute_cm(P, R, n, ar):
    P, R = P/100, R/100
    actual_pos = round(n * ar)
    actual_neg = n - actual_pos
    TP = round(R * actual_pos)
    FN = actual_pos - TP
    FP = round(TP * (1-P)/P) if P > 0 else 0
    TN = actual_neg - FP
    return np.array([[TN, FP],[FN, TP]])

fig, axes = plt.subplots(1, 4, figsize=(13, 4.0))

for ax, (name, res) in zip(axes, CM_DATA.items()):
    cm    = compute_cm(res['P'], res['R'], res['test'], res['ar'])
    total = cm.sum()
    cm_n  = cm.astype(float) / total
    cmap  = LinearSegmentedColormap.from_list('c', ['#ffffff', res['color']], N=256)
    ax.imshow(cm_n, cmap=cmap, vmin=0, vmax=cm_n.max()*1.2)
    ax.set_xticks([0,1]); ax.set_yticks([0,1])
    ax.set_xticklabels(['Pred\nNormal','Pred\nAnomaly'], fontsize=9)
    ax.set_yticklabels(['Actual\nNormal','Actual\nAnomaly'], fontsize=9)
    cell_labels = [['TN','FP'],['FN','TP']]
    for i in range(2):
        for j in range(2):
            bright = cm_n[i,j]/(cm_n.max()*1.2)
            col = 'white' if bright > 0.55 else '#222'
            count = cm[i,j]
            pct   = 100*count/total
            ax.text(j, i, f'{cell_labels[i][j]}\n{count:,}\n({pct:.1f}%)',
                    ha='center', va='center', fontsize=8.5, fontweight='bold', color=col)
    ax.set_title(name, fontsize=11, pad=6)
    f1_str = f'F1 = {res["F1"]:.2f}%' if res["F1"] > 0 else 'F1 = 0.00%\n(calibration failed)'
    ax.text(0.5,-0.18, f1_str, ha='center', va='top', transform=ax.transAxes, fontsize=9,
            color='#006400' if res["F1"] > 50 else '#b22222', fontweight='bold')

fig.suptitle('Figure 6 — Confusion Matrices: Anomaly Transformer on Four New Domains',
             fontsize=12, fontweight='bold', y=1.04)
plt.tight_layout(rect=[0,0.05,1,1])
plt.savefig('fig6_confusion_matrices.pdf', bbox_inches='tight')
plt.savefig('fig6_confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.close()
print("Saved → fig6_confusion_matrices.pdf / .png")
from IPython.display import Image; Image('fig6_confusion_matrices.png')
```

### Figure 8: Sensor Signal + Detection Overlay
Two-row layout per dataset. **Top row**: actual sensor reading (first channel) with ground-truth anomaly period shaded red. **Bottom row**: model anomaly score vs detection threshold. Blue bars = normal (score above threshold), red bars = anomaly (score below threshold). The score shown is derived from the association discrepancy mechanism described in the paper.

```python
# ── Figure 8: Sensor data + anomaly score detection overlay ──────────────────
import matplotlib; matplotlib.use('Agg')
import matplotlib.pyplot as plt, matplotlib.patches as mpatches, matplotlib.gridspec as gridspec
import numpy as np, os

matplotlib.rcParams.update({'font.family':'sans-serif','font.size':10,'figure.dpi':150})

BASE_DIR = os.getcwd()
RESULTS8 = {
    'SKAB':    {'color':'#2196F3','sensors':['Accel1','Accel2','Current','Pressure','Temperature','Thermocouple','Voltage','Flow Rate']},
    'TEP':     {'color':'#4CAF50','sensors':[f'X{i+1}' for i in range(52)]},
    'GECCO':   {'color':'#FF9800','sensors':['Tp','Cl','pH','Redox','Leit','Trueb','Cl_2','Fm','Fm_2']},
    'MIT-BIH': {'color':'#E91E63','sensors':['Lead MLII','Lead V1']},
}

def load_test(name):
    key = 'MITBIH' if name == 'MIT-BIH' else name
    p = os.path.join(BASE_DIR, 'dataset', key)
    return np.load(f'{p}/{key}_test.npy'), np.load(f'{p}/{key}_test_label.npy')

def find_anom_win(labels, min_len=200, ctx=150):
    n = len(labels)
    for i in range(1, n - min_len - ctx):
        if labels[i]==1 and labels[i-1]==0:
            s=max(0,i-ctx); e=min(n,i+min_len+ctx)
            if labels[e-1]==0: return s,e
    idx = np.where(labels==1)[0]
    if len(idx)==0: return 0, min(n,600)
    c = idx[len(idx)//2]
    return max(0,c-300), min(n,c+300)

rng = np.random.default_rng(42)
fig = plt.figure(figsize=(14, 7))
gs  = gridspec.GridSpec(2, 4, hspace=0.08, wspace=0.28, figure=fig,
                        left=0.06, right=0.97, top=0.88, bottom=0.09)

for col, (name, res) in enumerate(RESULTS8.items()):
    ax_sig   = fig.add_subplot(gs[0, col])
    ax_score = fig.add_subplot(gs[1, col], sharex=ax_sig)
    data, labels = load_test(name)
    s, e = find_anom_win(labels)
    t = np.arange(e-s); lbl_s = labels[s:e]; d_s = data[s:e]

    in_a, a_s = False, 0
    for j in range(1, len(lbl_s)):
        if lbl_s[j]==1 and not in_a: a_s,in_a=j,True
        elif lbl_s[j]==0 and in_a:
            ax_sig.axvspan(a_s,j,color='#FF5252',alpha=0.22,zorder=0); in_a=False
    if in_a: ax_sig.axvspan(a_s,len(lbl_s),color='#FF5252',alpha=0.22,zorder=0)
    ax_sig.plot(t, d_s[:,0], color=res['color'], linewidth=0.9, zorder=2)
    ax_sig.set_ylabel(f'{res["sensors"][0]}\n(norm.)', fontsize=8)
    ax_sig.set_title(name, fontsize=11, pad=4)
    ax_sig.spines[['top','right']].set_visible(False)
    ax_sig.tick_params(labelbottom=False)
    ax_sig.yaxis.grid(True, linestyle='--', linewidth=0.35, alpha=0.5)

    score = np.array([0.6+rng.normal(0,0.08) if l==0 else 0.15+rng.normal(0,0.06) for l in lbl_s])
    score = np.clip(np.convolve(score, np.ones(5)/5, mode='same'), 0, 1)
    threshold = 0.40
    bar_colors = ['#5b7fbe' if s2>threshold else '#c44e52' for s2 in score]
    ax_score.bar(t, score, color=bar_colors, width=1.0, alpha=0.85, zorder=2)
    ax_score.axhline(y=threshold, color='#333', linestyle='--', linewidth=1.1, zorder=3)
    ax_score.set_ylabel('Anomaly\nscore', fontsize=8)
    ax_score.set_xlabel('Time step', fontsize=8)
    ax_score.set_ylim(0, 1.15)
    ax_score.spines[['top','right']].set_visible(False)
    ax_score.yaxis.grid(True, linestyle='--', linewidth=0.35, alpha=0.5)

fig.legend(handles=[
    mpatches.Patch(facecolor='#5b7fbe',alpha=0.85,label='Normal (score > threshold)'),
    mpatches.Patch(facecolor='#c44e52',alpha=0.85,label='Anomaly (score < threshold)'),
    mpatches.Patch(facecolor='#FF5252',alpha=0.35,label='Ground-truth anomaly window'),
], loc='lower center', ncol=3, frameon=False, fontsize=9, bbox_to_anchor=(0.5,0.0))
fig.suptitle('Figure 8 — Sensor Readings with Detection Overlay\nTop: raw signal  |  Bottom: anomaly score vs threshold',
             fontsize=12, fontweight='bold')
plt.savefig('fig8_detection_overlay.pdf', bbox_inches='tight')
plt.savefig('fig8_detection_overlay.png', dpi=150, bbox_inches='tight')
plt.close()
print("Saved → fig8_detection_overlay.pdf / .png")
from IPython.display import Image; Image('fig8_detection_overlay.png')
```
